```python
import pandas as pd

# List of the CSV files you want to merge
# These are the filenames you provided from the upload
csv_files_to_merge = [
    "openlca_results_recorrected_calcsetup_ELCD.csv",
    "openlca_results_recorrected_calcsetup_NEEDS18.csv",
    "openlca_results_recorrected_calcsetup_Ozi.csv",
    "openlca_results_recorrected_calcsetup_USDA_2019.csv"
]

# Name for the output merged CSV file
output_merged_filename = "merged_openlca_results.csv"

# Create an empty list to hold the DataFrames
list_of_dataframes = []

print("Starting the merge process...")

# Loop through the list of CSV files
for filename in csv_files_to_merge:
    try:
        print(f"Reading file: {filename}...")
        # Read each CSV file into a DataFrame
        df = pd.read_csv(filename)
        list_of_dataframes.append(df)
        print(f"Successfully read {filename}, found {len(df)} rows.")
    except FileNotFoundError:
        print(f"Error: The file '{filename}' was not found. Please check the path.")
    except Exception as e:
        print(f"An error occurred while reading '{filename}': {e}")

# Check if any DataFrames were loaded
if list_of_dataframes:
    try:
        print("\nConcatenating DataFrames...")
        # Concatenate all the DataFrames in the list
        # ignore_index=True will re-index the new DataFrame from 0 to n-1
        merged_df = pd.concat(list_of_dataframes, ignore_index=True)
        print(f"Successfully concatenated all files. The merged DataFrame has {len(merged_df)} rows.")

        print(f"\nSaving the merged DataFrame to '{output_merged_filename}'...")
        # Save the merged DataFrame to a new CSV file
        # index=False means the DataFrame index will not be written into the CSV file
        merged_df.to_csv(output_merged_filename, index=False)
        print(f"Successfully saved the merged data to '{output_merged_filename}'.")
    except Exception as e:
        print(f"An error occurred during concatenation or saving: {e}")
else:
    print("\nNo data was loaded from the CSV files. Merged file not created.")

print("\nMerge process finished.")
```
