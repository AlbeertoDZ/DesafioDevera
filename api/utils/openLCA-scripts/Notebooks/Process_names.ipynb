```python
import olca_ipc as olca
import olca_schema
import os

# --- Configuration ---
OLCA_PORT = 8080
OUTPUT_FILENAME = "process_exiobase.txt"
# --- End of Configuration ---

def export_all_process_names():
    """
    Connects to OpenLCA, fetches all process names from the currently active database,
    and exports them to a text file, one name per line.
    """
    client = None
    process_names_count = 0
    print(f"🚀 Starting script to export all process names to '{OUTPUT_FILENAME}'...")

    try:
        print(f"🔌 Attempting to connect to OpenLCA on port {OLCA_PORT}...")
        client = olca.Client(OLCA_PORT)
        print("✅ Successfully connected to OpenLCA.")
        
        active_db_info = client.get_database_info()
        if active_db_info and hasattr(active_db_info, 'name') and active_db_info.name:
            print(f"ℹ️ Connected to active database: '{active_db_info.name}'")
        else:
            print("⚠️ Could not retrieve active database name, but proceeding.")

        print("ℹ️ Fetching all process descriptors from the active database...")
        # Get descriptors, which are more lightweight than full process objects
        all_process_descriptors = client.get_descriptors(olca_schema.Process)
        
        if not all_process_descriptors:
            print("❌ No process descriptors found in the active database, or failed to fetch.")
            return

        process_names = [desc.name for desc in all_process_descriptors if hasattr(desc, 'name') and desc.name]
        process_names_count = len(process_names)
        
        if not process_names:
            print("🤷 No process names extracted from descriptors (perhaps descriptors have no names or list is empty).")
            return

        print(f"👍 Found {process_names_count} process names.")

        # Sort names alphabetically (optional, but good for consistency)
        process_names.sort()

        print(f"💾 Saving process names to '{OUTPUT_FILENAME}'...")
        with open(OUTPUT_FILENAME, 'w', encoding='utf-8') as f:
            for name in process_names:
                f.write(name + "\n")
        
        print(f"🎉 Successfully saved {process_names_count} process names to '{OUTPUT_FILENAME}'.")

    except ConnectionRefusedError:
        print(f"❌ Connection refused. Please ensure the OpenLCA application is running and the IPC server is enabled on port {OLCA_PORT}.")
    except Exception as e:
        print(f"💥 An unexpected error occurred: {e}")
        import traceback
        traceback.print_exc()
    finally:
        if client:
            print("🔒 Closing connection to OpenLCA.")
            # client.close() # olca-ipc client typically doesn't have an explicit close method; connections are managed per request.
        print("🏁 Script finished.")

if __name__ == "__main__":
    export_all_process_names()
```

```python
import olca_ipc as olca
import olca_schema
import os

# --- Configuration ---
OLCA_PORT = 8080
OUTPUT_FILENAME = "all_exiobase_process_details.txt" # Changed filename
# --- End of Configuration ---

def export_exiobase_process_details():
    """
    Connects to OpenLCA, fetches details for all processes from the
    currently active Exiobase database (name, category, location),
    and exports them to a text file.
    """
    client = None
    process_details_list = []
    print(f"🚀 Starting script to export Exiobase process details to '{OUTPUT_FILENAME}'...")

    try:
        print(f"🔌 Attempting to connect to OpenLCA on port {OLCA_PORT}...")
        client = olca.Client(OLCA_PORT)
        print("✅ Successfully connected to OpenLCA.")
        print("ℹ️ Assuming connection to the currently active database in OpenLCA (should be Exiobase).")

        print("ℹ️ Fetching all process descriptors from the active database...")
        all_process_descriptors = client.get_descriptors(olca_schema.Process)
        
        if not all_process_descriptors:
            print("❌ No process descriptors found in the active database, or failed to fetch.")
            return

        print(f"👍 Found {len(all_process_descriptors)} process descriptors. Extracting details...")

        for desc_idx, desc in enumerate(all_process_descriptors):
            if not (hasattr(desc, 'name') and desc.name):
                print(f"⚠️ Descriptor {desc_idx} has no name, skipping.")
                continue

            process_name = desc.name
            category_path = desc.category if hasattr(desc, 'category') and desc.category else "N/A"
            location_code = "N/A"
            location_name = "N/A" # Optional: fetch full location name if needed, but code is often sufficient

            if hasattr(desc, 'location') and desc.location and hasattr(desc.location, 'code') and desc.location.code:
                location_code = desc.location.code
            # If you also want the full location name (more IPC calls, can be slow for many processes):
            # if hasattr(desc, 'location') and desc.location and hasattr(desc.location, 'id'):
            #     try:
            #         loc_obj = client.get(olca_schema.Location, desc.location.id)
            #         if loc_obj and hasattr(loc_obj, 'name'):
            #             location_name = loc_obj.name
            #     except Exception as e_loc:
            #         print(f"Error fetching location name for {desc.location.id}: {e_loc}")


            # Create a more unique identifier. You can adjust the format.
            # Example 1: Name (Location Code) [Category Path]
            # unique_identifier = f"{process_name} ({location_code}) [{category_path}]"
            # Example 2: Using a separator, more machine-readable
            unique_identifier = f"NAME={process_name};LOC={location_code};CAT={category_path};ID={desc.id}"


            process_details_list.append(unique_identifier)
            
            if (desc_idx + 1) % 500 == 0: # Print progress
                print(f"    Processed {desc_idx + 1}/{len(all_process_descriptors)} descriptors...")


        if not process_details_list:
            print("🤷 No process details extracted.")
            return

        print(f"👍 Extracted details for {len(process_details_list)} processes.")

        # Sort identifiers (optional)
        process_details_list.sort()

        print(f"💾 Saving process details to '{OUTPUT_FILENAME}'...")
        with open(OUTPUT_FILENAME, 'w', encoding='utf-8') as f:
            for detail_line in process_details_list:
                f.write(detail_line + "\n")
        
        print(f"🎉 Successfully saved {len(process_details_list)} process details to '{OUTPUT_FILENAME}'.")

    except ConnectionRefusedError:
        print(f"❌ Connection refused. Please ensure the OpenLCA application is running and the IPC server is enabled on port {OLCA_PORT}.")
    except Exception as e:
        print(f"💥 An unexpected error occurred: {e}")
        import traceback
        traceback.print_exc()
    finally:
        if client:
            print("🔒 IPC client session will implicitly close.")
        print("🏁 Script finished.")

if __name__ == "__main__":
    export_exiobase_process_details()
```

```python
import olca_ipc as olca
import olca_schema # Ensure olca_schema is available if you use olca_schema.Process
import os
import traceback

# --- Configuration ---
OLCA_PORT = 8080
OUTPUT_FILENAME = "process_exiobase_unique_names.txt"
# --- End of Configuration ---

def export_unique_process_names():
    """
    Connects to OpenLCA, fetches all process names from the currently active database,
    ensures only unique names are kept, sorts them, and exports them to a text file,
    one name per line.
    """
    client = None
    original_names_count = 0 # Renamed from process_names_count for clarity
    unique_process_names_count = 0
    print(f"🚀 Starting script to export unique process names to '{OUTPUT_FILENAME}'...")

    try:
        print(f"🔌 Attempting to connect to OpenLCA on port {OLCA_PORT}...")
        client = olca.Client(OLCA_PORT)
        print("✅ Successfully connected to OpenLCA.")
        
        # The following section caused the error and is not strictly necessary
        # for fetching process descriptors from the active database.
        # It's commented out to resolve the AttributeError.
        # --- Commented out section ---
        # active_db_info = client.get_database_info() # This line caused the error
        # if active_db_info and hasattr(active_db_info, 'name') and active_db_info.name:
        # print(f"ℹ️ Operating on the active database (name retrieval skipped).")
        # else:
        #     print("⚠️ Could not retrieve active database name (method not available), but proceeding.")
        # --- End of commented out section ---
        print("ℹ️ Will fetch processes from the currently active database in OpenLCA.")

        print("ℹ️ Fetching all process descriptors from the active database...")
        # Ensure olca_schema.Process is correctly referenced.
        # If olca_schema is not directly imported but part of olca.schema, it might be:
        # all_process_descriptors = client.get_descriptors(olca.schema.Process)
        # However, 'import olca_schema' usually implies it's a separate module part of the install.
        all_process_descriptors = client.get_descriptors(olca_schema.Process)
        
        if not all_process_descriptors:
            print("❌ No process descriptors found in the active database, or failed to fetch.")
            return

        all_names_from_descriptors = [
            desc.name for desc in all_process_descriptors if hasattr(desc, 'name') and desc.name
        ]
        original_names_count = len(all_names_from_descriptors)
        print(f"👍 Found {original_names_count} process names (including potential duplicates from descriptors).")

        if not all_names_from_descriptors:
            print("🤷 No process names extracted from descriptors.")
            return

        unique_process_names = sorted(list(set(all_names_from_descriptors)))
        unique_process_names_count = len(unique_process_names)
        
        if not unique_process_names:
            print("🤷 No unique process names to save (list is empty after attempting to find uniques).")
            return
            
        print(f"☑️ Identified {unique_process_names_count} unique process names.")

        print(f"💾 Saving unique process names to '{OUTPUT_FILENAME}'...")
        with open(OUTPUT_FILENAME, 'w', encoding='utf-8') as f:
            for name in unique_process_names:
                f.write(name + "\n")
        
        print(f"🎉 Successfully saved {unique_process_names_count} unique process names to '{OUTPUT_FILENAME}'.")

    except ConnectionRefusedError:
        print(f"❌ Connection refused. Please ensure the OpenLCA application is running and the IPC server is enabled on port {OLCA_PORT}.")
    except AttributeError as ae:
        print(f"💥 AttributeError occurred: {ae}")
        print("   This might indicate an issue with object properties or methods in the olca-ipc library version.")
        traceback.print_exc()
    except Exception as e:
        print(f"💥 An unexpected error occurred: {e}")
        traceback.print_exc()
    finally:
        if client:
            print("🔒 IPC client session is typically managed per request by olca-ipc; no explicit client.close() is usually needed.")
        print("🏁 Script finished.")

if __name__ == "__main__":
    export_unique_process_names()
```
