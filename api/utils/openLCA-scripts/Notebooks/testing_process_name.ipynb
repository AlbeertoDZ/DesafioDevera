```python
import olca_ipc as olca
import olca_schema
import os

# --- Configuration ---
OLCA_PORT = 8080
OUTPUT_FILENAME = "process_exiobase.txt"
# --- End of Configuration ---

def export_all_process_names():
    """
    Connects to OpenLCA, fetches all process names from the currently active database,
    and exports them to a text file, one name per line.
    """
    client = None
    process_names_count = 0
    print(f"🚀 Starting script to export all process names to '{OUTPUT_FILENAME}'...")

    try:
        print(f"🔌 Attempting to connect to OpenLCA on port {OLCA_PORT}...")
        client = olca.Client(OLCA_PORT)
        print("✅ Successfully connected to OpenLCA.")
        
        active_db_info = client.get_database_info()
        if active_db_info and hasattr(active_db_info, 'name') and active_db_info.name:
            print(f"ℹ️ Connected to active database: '{active_db_info.name}'")
        else:
            print("⚠️ Could not retrieve active database name, but proceeding.")

        print("ℹ️ Fetching all process descriptors from the active database...")
        # Get descriptors, which are more lightweight than full process objects
        all_process_descriptors = client.get_descriptors(olca_schema.Process)
        
        if not all_process_descriptors:
            print("❌ No process descriptors found in the active database, or failed to fetch.")
            return

        process_names = [desc.name for desc in all_process_descriptors if hasattr(desc, 'name') and desc.name]
        process_names_count = len(process_names)
        
        if not process_names:
            print("🤷 No process names extracted from descriptors (perhaps descriptors have no names or list is empty).")
            return

        print(f"👍 Found {process_names_count} process names.")

        # Sort names alphabetically (optional, but good for consistency)
        process_names.sort()

        print(f"💾 Saving process names to '{OUTPUT_FILENAME}'...")
        with open(OUTPUT_FILENAME, 'w', encoding='utf-8') as f:
            for name in process_names:
                f.write(name + "\n")
        
        print(f"🎉 Successfully saved {process_names_count} process names to '{OUTPUT_FILENAME}'.")

    except ConnectionRefusedError:
        print(f"❌ Connection refused. Please ensure the OpenLCA application is running and the IPC server is enabled on port {OLCA_PORT}.")
    except Exception as e:
        print(f"💥 An unexpected error occurred: {e}")
        import traceback
        traceback.print_exc()
    finally:
        if client:
            print("🔒 Closing connection to OpenLCA.")
            # client.close() # olca-ipc client typically doesn't have an explicit close method; connections are managed per request.
        print("🏁 Script finished.")

if __name__ == "__main__":
    export_all_process_names()
```
